```python
from statistics import correlation

import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books =pd.read_csv(f"{path}/books.csv")
```

```python
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=True, ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing Values")

plt.show()
```

```python
import numpy as np

books["Missing Description"] = np.where(books["description"].isna(),1,0)
books["Age of Books"] = 2025 - books["published_year"]
# print(books["Age of Books"])
```

```python
column_of_interest = ["num_pages", "Age of Books","Missing Description","average_rating"]
correlation_matrix = books[column_of_interest].corr(method="spearman")

sns.set_theme(style="white")
plt.figure(figsize = (6,6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt = ".2f", cmap="coolwarm", cbar_kws={"label": "Spearman Correlation"})
heatmap.set_title("Correlation heatmap")
plt.show()
```

```python
books_missing = books[~(books["description"].isna()) &
      (~books["num_pages"].isna()) &
      (~books["average_rating"].isna()) &
      (~books["published_year"].isna())
]
```

```python
books_missing
```

```python
books_missing["categories"].value_counts().reset_index(name="count").sort_values("count", ascending=False)
```

```python
books_missing = books_missing.copy()
books_missing["words in description"] = books_missing["description"].str.split().str.len()
```

```python
books_missing_25words = books_missing[books_missing["words in description"]>=25]
```

```python
books_missing_25words
```

```python
books_missing_25words.loc[:,"title_and_subtitle"] = (
    np.where(books_missing_25words["subtitle"].isna(), books_missing_25words["title"],
             books_missing_25words[["title", "subtitle"]].astype(str).agg(": ".join,axis=1))
)
```

```python
books_missing_25words
```

```python
books_missing_25words.loc[:,"tagged_description"] = books_missing_25words[["isbn13","description"]].astype(str).agg(" ".join,axis=1)
```

```python
(
    books_missing_25words
    .drop(["subtitle","Missing Description","Age of Books","words in description"], axis=1).to_csv("books_cleaned.csv",index=False)
)
```
